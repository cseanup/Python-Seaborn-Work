```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style = 'darkgrid')
```

```python
tips = sns.load_dataset('tips')
tips.head()
tips.tail()
```

```python
# Relational Plot
```

```python
sns.relplot(x= 'total_bill', y ='tip', data  = tips)
```

```python
dir(sns.FacetGrid)
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, hue = 'smoker')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, hue = 'size')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, hue = 'sex',    palette='ch:r = 0.7, l = 0.3' )
```

```python
tips['sex'].value_counts()
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, hue = 'size')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, size = 'size') #Size of the objects
sns.relplot(x= 'total_bill', y = 'tip', data = tips, hue = 'size') # Change color of the objects
sns.relplot(x= 'total_bill', y = 'tip', data = tips, style = 'size') #Change style of the objects
```

```python
# modifiying the size of the object
sns.relplot(x= 'total_bill', y = 'tip', data = tips, size = 'size', sizes = (15,200))
```
